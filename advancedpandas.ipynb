**Q1 **Dataset Exploration

Perform the following exploratory steps:

Display the first 10 rows

Display the dataset structure using info()

Find the number of rows and columns

Identify the data types of each column

```python
import pandas as pd

# Load the dataset (Make sure the file path matches your environment)
df = pd.read_csv('Advanced_pandas_sales_dataset.csv')

# 1. Display the first 10 rows
print("--- First 10 Rows ---")
print(df.head(10))

# 2. Display the dataset structure
print("\n--- Dataset Structure ---")
df.info()

# 3. Find the number of rows and columns
print("\n--- Rows and Columns ---")
print(f"Shape of dataset: {df.shape}")

# 4. Identify the data types of each column
print("\n--- Data Types ---")
print(df.dtypes)
```

**Q2** Sales Filtering Analysis

Filter the dataset to display:

Orders where Sales > 3500

Orders where Region = "East" and Sales > 2500

Explain what insights you observe.


```python
# 1. Orders where Sales > 3500
high_sales = df[df['Sales'] > 3500]
print("--- Orders with Sales > 3500 ---")
print(high_sales)

# 2. Orders where Region = "East" and Sales > 2500
east_high_sales = df[(df['Region'] == 'East') & (df['Sales'] > 2500)]
print("\n--- East Region Orders with Sales > 2500 ---")
print(east_high_sales)
```

**Q3** Customer Purchase Analysis

Identify the top 5 customers based on total sales.

Hint: Use GroupBy with aggregation.

```python
# Top 5 customers based on total sales
top_customers = df.groupby('Customer')['Sales'].sum().reset_index()
top_customers = top_customers.sort_values(by='Sales', ascending=False).head(5)

print("--- Top 5 Customers by Total Sales ---")
print(top_customers)
```

**Q4** Regional Sales Performance

Calculate the total sales generated by each region.

Then identify:

Which region generated the highest revenue?

```python
# Total sales generated by each region
regional_sales = df.groupby('Region')['Sales'].sum().reset_index()
print("--- Total Sales by Region ---")
print(regional_sales)

# Identify highest revenue region
highest_region = regional_sales.sort_values(by='Sales', ascending=False).iloc[0]
print(f"\n Highest revenue generated by: {highest_region['Region']} (${highest_region['Sales']:,.2f})")
```

**Q5** Product Performance Analysis

Calculate the total quantity sold for each product.

Which product appears to be the most popular?



```python
# Total quantity sold for each product
product_qty = df.groupby('Product')['Quantity'].sum().reset_index()
print("--- Total Quantity Sold by Product ---")
print(product_qty)

# Most popular product
most_popular = product_qty.sort_values(by='Quantity', ascending=False).iloc[0]
print(f"\n Most popular product: {most_popular['Product']} ({most_popular['Quantity']} units sold)")
```

**Q6 **Multi-Level Grouping

Group the dataset by: Region and Category
Calculate the total sales for each combination.

What insights can you derive from this analysis?

```python
# Group by Region and Category
multi_group = df.groupby(['Region', 'Category'])['Sales'].sum().reset_index()
print("--- Total Sales by Region and Category ---")
print(multi_group)
```

**Q7** Pivot Table Creation

Create a pivot table showing:

Rows → Region

Columns → Product

Values → Total Sales

Which region sells the most laptops?



```python
# Create pivot table
pivot_sales = df.pivot_table(values='Sales', index='Region', columns='Product', aggfunc='sum', fill_value=0)
print("--- Regional Product Sales Pivot Table ---")
print(pivot_sales)

# Check which region sells the most laptops (Assumes column name is 'Laptop')
if 'Laptop' in pivot_sales.columns:
    top_laptop_region = pivot_sales['Laptop'].idxmax()
    print(f"\n Region that sells the most laptops: {top_laptop_region}")
else:
    print("\n Column 'Laptop' not found. Check your specific product names in the dataset.")
```

**Q8** Profitability Analysis

Calculate the average profit for each product category.

Which category is most profitable?

```python
# Average profit for each product category
avg_profit = df.groupby('Category')['Profit'].mean().reset_index()
print("--- Average Profit by Category ---")
print(avg_profit)

# Most profitable category
most_profitable = avg_profit.sort_values(by='Profit', ascending=False).iloc[0]
print(f"\n Most profitable category: {most_profitable['Category']}")
```

**Q9** Discount Impact Study

Analyze the relationship between discount and profit.

Tasks:

Calculate the average profit for each discount level

Identify whether higher discounts lead to lower profit

```python
# 1. Average profit for each discount level
discount_impact = df.groupby('Discount')['Profit'].mean().reset_index()
print("--- Average Profit by Discount Level ---")
print(discount_impact)
```

**Q10** End-to-End Data Wrangling Task

Perform the following workflow:

Filter the dataset where Sales > 3000

Group data by Region and Product

Calculate total sales

Reshape the result into a pivot table

Explain the business insights obtained from this analysis

```python
# Step 1: Filter dataset where Sales > 3000
wrangled_df = df[df['Sales'] > 3000]

# Step 2 & 3: Group by Region and Product, calculate total sales
grouped_wrangled = wrangled_df.groupby(['Region', 'Product'])['Sales'].sum().reset_index()

# Step 4: Reshape the result into a pivot table
final_pivot = grouped_wrangled.pivot_table(values='Sales', index='Region', columns='Product', fill_value=0)

print("--- Final Wrangled Pivot Table (Sales > 3000) ---")
print(final_pivot)
```
